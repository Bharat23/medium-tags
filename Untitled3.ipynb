```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv('Data_1_1000.csv')

split = 0.70
split_index = int(len(data)*split);

train = data[:split_index]
test = data[split_index:]
train.head()
```

```
tag_cols = train.iloc[:, 14:-1]
content_col = train.iloc[:, -1]
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
import re

# filter and remove numbers
filtered_content_col = []
for i in range(len(content_col)):
  filtered_content_col.append(re.sub(r'[\d\_]+', '', content_col[i]))



# generate TFIDF vectorizer
vectorized = TfidfVectorizer(stop_words='english', min_df = 0.01)
d = vectorized.fit_transform(filtered_content_col)

# create a dataframe of the matrix
vocab_dataframe = pd.DataFrame(d.todense(), index=[x for x in range(0, len(filtered_content_col))], columns=vectorized.get_feature_names())

print(vocab_dataframe)
```

```
# try to find clusters based on word usage

from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=3).fit(vocab_dataframe)
print(k_means.labels_)

a = np.array(tag_cols[1:2]).flatten()
b = np.array(tag_cols[2:3]).flatten()

count = 0
for i in range(len(a)):
  if a[i] == b[i] and a[i] != 0:
    count += 1
print(count)

# print(np.array(a[0:1]).flatten())
```

```
import matplotlib.pyplot as plt 

plt.plot(vocab_dataframe)
plt.legend()
plt.show()
```
